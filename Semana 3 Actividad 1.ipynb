**1. FUNDAMENTOS DE BASES DE DATOS Y ALMACENES DE DATOS**

Se entiende por base de datos a la colección de datos almacenados, ordenados y estructurados, con el fin de poder obtener acceso a estos de forma rápida y segura para brindar respuestas a las necesidades de información de una organización.
Existen las bases de datos relacionales y las no relacionales, la primera almacena la información en tablas las cuales pueden estar conectadas unas a otras a través de relaciones y pueden ser accedidas por medio de un lenguaje de consulta, como SQL, estas son base importante para los científicos de datos ya que a partir de estas relaciones se puede obtener un mejor entendimiento de los datos. Las no relacionales por el contrario no almacenan la información en tablas ni son fácilmente accedidas por algún lenguaje, no hay orden en los datos y no están estructurados.

Un Data Warehouse Es una arquitectura de almacenamiento de datos, que puede ser alimentada por diferentes bases de datos relacionales, sistemas transaccionales ,entre otros, su principal característica es que los datos que ingresan son tratados y preparados para futuras consultas con el objetivo de permitir a las organizaciones analizar, comprender y utilizar sus datos para toma de decisiones estratégicas. 


```python
# 2. SELECCIÓN Y LIMPIEZA DE LOS  DATOS EN PYTHON


# cargue de la informacion a analizar y limpiar
input="https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
```

```python
# Importamos la libreria panda para cague de los datos
import pandas as pd

# cargue de la informacion en dataframe llamado DF_BASE
DF_BASE = pd.read_csv(input)
DF_BASE
```

```python
#Verificacion si en la informacion falta algun dato.
DF_BASE.isnull().values.any() 
# se puede verificar tambien por columnas, resultante true si falta o false sino falta para dicha columna
DF_BASE.isnull().any()  

#Alternativa a isnull
DF_BASE.isna().values.any()

#Alternativa a isnull por columnas
DF_BASE.isna().any()
```

```python
#Solucion 1: 

#1.1  Descartar filas con valores faltantes
# Se copia el data frame en uno temporal para no perder el original en fase de prueba de soluciones
DF_SOL1 = DF_BASE.copy()
DF_SOL1.dropna(inplace=True)  # se indica cambiar el data frame
DF_SOL1.isna().values.any()  # verificación que se eliminaron las filas que conteniana NAN o NAT
DF_SOL1 # Se obtienen 29958 Filas, hay gran perdida de registros

#1.2 Descartar columnas con valores faltantes
DF_SOL12 = DF_BASE.copy()
DF_SOL12.dropna(axis = 1,inplace=True) # Axis 1 indica elimiancion por colunmas donde exista por lo menos un NaN NaT
DF_SOL12  # Se obtiene solo dos columnas, gran perdida de informacion

#1.3 Eliminar filas donde todos los elementos son NaN NaT
DF_SOL13 = DF_BASE.copy()
DF_SOL13.dropna(how='all', inplace=True)  # se indica all para condicion de todos los elemntos de la fila n
DF_SOL13  # NO  existe filas donde todos los registros sean NaN

#1.4 Eliminar filas que por lo menos tenga dos valores diferentes a NaN
DF_SOL14 = DF_BASE.copy()
DF_SOL14.dropna(thresh = 4, inplace=True) # thresh indica ell numero de valores minimos para no eliminar
DF_SOL14  # no hay filas con nulos y al menos 4 valores

#1.5  Definir las columnas donde buscar NaN indicando numero d valores minimos
DF_SOL15 = DF_BASE.copy()
DF_SOL15.dropna(thresh = 1,axis = 1, inplace=True)
DF_SOL15 # no hay columnas con nulos y al menos 5 valores NaN

```

```python
#Solución 2:

DF_SOL2 = DF_BASE.copy()

#Obtener la media de una columna EJE, X2 para reemplazar los NaN con la media
media = DF_SOL2.X2.mean()
DF_SOL2['X2'].fillna(value = media, inplace = True)
DF_SOL2

# #Obtener la moda de una columna EJE, X2 para reemplazar los NaN con la moda
DF_SOL21 = DF_BASE.copy()
moda = DF_SOL21.X5.mode()
moda

DF_SOL21['X5'].fillna(value = moda[0], inplace = True)
DF_SOL21 

#LIMPIAR SOLO COLUMNAS ESPECIFICAS
DF_SOL22 = DF_BASE.copy()
DF_SOL22.dropna(subset=['X2','X3','X4','X5'], inplace = True)
DF_SOL22 # Se eliminan 5 registros donde se cumple la condicion de NaN en las columnas especificadas


DF_SOL23 = DF_BASE.copy()
DF_SOL23.X5.mode()[0]

# Selecicon de criterios favoritos para reemplazar, por ejemplo se toma la columna X5  para reemplaz los NaN por la moda
# y se toma la columna X3 para reemplazar los NaN por la media de esta
favs = {'X5': DF_SOL23.X5.mode()[0], 'X3': DF_SOL23['X3'].mean()}
DF_SOL23.X5.fillna(DF_SOL23.X5.mode()[0], inplace = True)
DF_SOL23.X3.fillna(DF_SOL23.X3.mean(), inplace = True)
DF_SOL23
```
