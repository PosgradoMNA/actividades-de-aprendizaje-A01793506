1. Eliminar el valor perdido (Puede ser eliminando la fila que contiene el valor perdido.
2. Eliminando la variable.

Ambas opciones son recomendadas cuando no se tiene mucha información en las demas variables o esta por si sola no es de gran importancia.

3. Reemplazar los valores. (Por ejemplo con el promedio de toda la variable.
para el caso de varibles categoricas se puede usar la Moda.

4. Se puede reemplzar tambien basados en otras funciones de la variable.

5. Por ultimo, se puede dejar la variable como dato faltante ya que podria resultar iimportante.
