```python
#Objetos y clases
#Cada tipo de datos es representado como un objeto, es decir este es una instancia del tipo de dato
#Metdos son las clases o funciones que utlizamos para interactuar cob los objetos

#Creacion de clases
#Atributos de datos
#Metodos

#Creando una clase
#Clase para instanciar productos y valor del producto
class productos(object):
 def __init__(self,producto,precio):
    self.producto = producto;
    self.precio = precio;

    #agregamos metodos para cambiar valores de los datos
 def cambia_precio(self, nv):
    self.precio = nv  

#probando la clase
canasta1 = productos('Manzanas',5000)  
canasta1.precio
canasta1.cambia_precio(7000)  # probando metodo para cambiar precio
canasta1.precio



```
