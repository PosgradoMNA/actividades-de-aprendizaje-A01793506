```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

##1. Obtención de datos y cargue en un data frame.

```python
url = "https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
df = pd.read_csv(url, index_col=0)
df.head()
```

##2. Obten la información del DataFrame con los métodos y propiedades: shape, columns, head(), dtypes, info(), isna()

```python
#obtenemos las dimensiones del data frame
print('Dimensiones dataframe:', df.shape, '\n')
#Obtenemos los nombres de las columnas
print(df.columns)
#Obtenemos los tipos de datos
print(df.dtypes)
#Obtenemos una infomracion mas completa a traves del metodo Info acerca de las caracteristicas del DF
print(df.info())
#verificamos valores NaN por columna, verificamos que solo la columna 1 no tiene valores NaN
df.isna().any()
```

##3. Limpia los datos eliminando los registros nulos o rellena con la media de la columna

```python
df1 = df.copy()
df1.dropna(how='all', inplace=True)  # se indica all para condicion de todos los elemntos de la fila n
df1  # NO  existe filas donde todos los registros sean NaN
# por lo tanto se procede a transformar datos con la media
```

```python
#Comprobamos el porcentaje de registros nulos en el dataframe
cantnull = df[df.isnull().any(axis=1)].shape[0]
print('Cantidad de registros NaN', cantnull)
#Se obtienen 42 registros nulos entre los 3000, veamos el porcentaje que representa
porc = (cantnull / df.shape[0])*100
print('Porcentaje de registros NaN', porc)

#Observamos que la cantidad de regsitos nulos es 14%, es una cifra significativa, por lo tanto procedemos
#a imputar con la media para los variables numericas y con la moda para las varibales categoricas.

#De acuerdo a la descripción de las columnas, las numericas son x1, x5 de x12 a x23

df1['X5'].fillna(value = df1.X5.mean(), inplace = True)
df1['X12'].fillna(value = df1.X12.mean(), inplace = True)
df1['X13'].fillna(value = df1.X13.mean(), inplace = True)
df1['X14'].fillna(value = df1.X14.mean(), inplace = True)
df1['X15'].fillna(value = df1.X15.mean(), inplace = True)
df1['X16'].fillna(value = df1.X16.mean(), inplace = True)
df1['X17'].fillna(value = df1.X17.mean(), inplace = True)
df1['X18'].fillna(value = df1.X18.mean(), inplace = True)
df1['X19'].fillna(value = df1.X19.mean(), inplace = True)
df1['X20'].fillna(value = df1.X20.mean(), inplace = True)
df1['X21'].fillna(value = df1.X21.mean(), inplace = True)
df1['X22'].fillna(value = df1.X22.mean(), inplace = True)
df1['X23'].fillna(value = df1.X23.mean(), inplace = True)
df1['Y'].fillna(value = df1.Y.mean(), inplace = True)

#Eliminamos valores nulos de las columnas categoricas
df1.dropna(subset=["X2"], axis=0, inplace=True)
df1.dropna(subset=["X3"], axis=0, inplace=True)
df1.dropna(subset=["X4"], axis=0, inplace=True)
df1.dropna(subset=["X6"], axis=0, inplace=True)
df1.dropna(subset=["X7"], axis=0, inplace=True)
df1.dropna(subset=["X8"], axis=0, inplace=True)
df1.dropna(subset=["X9"], axis=0, inplace=True)
df1.dropna(subset=["X10"], axis=0, inplace=True)
df1.dropna(subset=["X11"], axis=0, inplace=True)
df1.reset_index(drop=True, inplace=True)

#'por ultimo comprobamos que no  quedan registos NaN en el data frame'
df1.isna().any()
```

```python
#Como habiamos revisado en el punto 2 y 4, se considera necesario escalar los datos debido 
#la alta dispersion de la informaciom entre las columnas

Scaler = StandardScaler()
df2 = Scaler.fit_transform(df1)
df3=pd.DataFrame(df2, columns=df1.columns)
df3.head()

#Podemos observar que hay una distribucion normal en los datos posterior al escalamiento.
```

##7. Reduce las dimensiones con PCA, si consideras necesario.
Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
Indica la importancia de las variables en cada componente

```python
Comp = PCA()
resul_pca= Comp.fit_transform(df3)
resul_pca
```

```python
df4 = pd.DataFrame({'% Varianza del componente': np.round(Comp.explained_variance_ratio_,2) * 100, 
                                        '% Varianza acumulada': np.cumsum(Comp.explained_variance_ratio_) * 100
})
df4
```

De acuerdo a las condiciones del ejercicio estariamos cubriendo el 70% de la varianza del conjunto de datos hasta el componente 7, por lo tanto se aprecia que etariamos haciendo una reducción importante de la dimensionalidad de los datos.

```python


dfpca = pd.DataFrame(resul_pca)
df4 = pd.DataFrame(Comp.components_.round(4), columns = dfpca.columns, index = df3.columns)
df4.iloc[:,:]
df4.iloc[:,:8].abs().idxmax()
```

Podemos obsevar que los componentes (los 7 que aportan mas del 70%) que mas aportan en el modelo para la precisión de los datos son los componentes 1 y 2, con el mayor aporte para la variable X5, los componentes 3 y 4, con la mayoria de aporte para la varibale X14.

## 9. Realiza la visualización de los datos usando por lo menos 3 gráficos que consideres adecuados: plot, scatter, jointplot, boxplot, areaplot, pie chart, pairplot, bar chart, etc.

```python
#Uso de Scree plot para observar la importancia de las caracteristicas
PC_components = np.arange(Comp.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = Comp.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(Comp.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```
